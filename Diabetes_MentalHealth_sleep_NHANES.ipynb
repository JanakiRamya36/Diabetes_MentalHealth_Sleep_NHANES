# Do different Categories related to depression and sleep affect the levels of HbA1c

# One-way ANOVA

```python
from scipy.stats import f_oneway

f_n=join[join['Feeling Depressed']=='no']['HbA1c']
f_h=join[join['Feeling Depressed']=='> half days']['HbA1c']
f_s=join[join['Feeling Depressed']=='several days']['HbA1c']
f_e=join[join['Feeling Depressed']=='every day']['HbA1c']

result1=f_oneway(f_n,f_h,f_s,f_e)
print(f'ANOVA result: Effect of Depression on HbA1c levels:\n{result1}')
```

```python
from scipy.stats import f_oneway

t_n=join[join['Trouble sleeping/ Sleeping too much ']=='no']['HbA1c']
t_h=join[join['Trouble sleeping/ Sleeping too much ']=='> half days']['HbA1c']
t_s=join[join['Trouble sleeping/ Sleeping too much ']=='several days']['HbA1c']
t_e=join[join['Trouble sleeping/ Sleeping too much ']=='every day']['HbA1c']

result2=f_oneway(t_n,t_h,t_s,t_e)
print(f'ANOVA result: Effect of Sleep disturbances on HbA1c levels:\n{result2}')
```

```python
from scipy.stats import f_oneway

p_n=join[join['Poor appetite/ overeating']=='no']['HbA1c']
p_h=join[join['Poor appetite/ overeating']=='> half days']['HbA1c']
p_s=join[join['Poor appetite/ overeating']=='several days']['HbA1c']
p_e=join[join['Poor appetite/ overeating']=='every day']['HbA1c']

result3=f_oneway(p_n,p_h,p_s,p_e)
print(f'ANOVA result: Effect of Appetite on HbA1c levels:\n{result3}')
```

```python
from scipy.stats import f_oneway

e_n=join[join['Feeling tired/little energy']=='no']['HbA1c']
e_h=join[join['Feeling tired/little energy']=='> half days']['HbA1c']
e_s=join[join['Feeling tired/little energy']=='several days']['HbA1c']
e_e=join[join['Feeling tired/little energy']=='every day']['HbA1c']

result4=f_oneway(e_n,e_h,e_s,e_e)
print(f'ANOVA result: Effect of Energy levels on HbA1c levels:\n{result4}')
```

# Association between Sleep on week days and HbA1c

```python
x=stats.pearsonr(join['HbA1c'],join['Sleep- weekdays'])
print(f'Correlation between sleep on weekdays and HbA1c:\n{x}')
print('Spearman correlation: ',join['HbA1c'].corr(join['Sleep- weekdays'],'spearman'))
```

```python
x = join['Sleep- weekdays']
y = join['HbA1c']
plt.scatter(x, y, label='Data')

coefficients = np.polyfit(x, y, 1)
poly = np.poly1d(coefficients)
plt.plot(x, poly(x), color='red', label='Regression Line')

plt.xlabel('Sleep- weekdays')
plt.ylabel('HbA1c')
plt.title('Regression Plot: Sleep- weekdays vs. HbA1c')
plt.legend()
plt.show()
```
